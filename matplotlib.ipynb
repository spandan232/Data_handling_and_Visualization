```python
import matplotlib.pyplot as plt
x=[1,2,3,4]
y=[4,5,6,7]
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.title("Line graph")
plt.plot(x,y)
plt.show()
```

```python

x=[20,30,40,50]

plt.pie(x)
plt.show()
```

```python

#Pie chart with labels
y=[10,20,30,40]
mylabels=["Apples","Bananas","Cherries","Dates"]
plt.pie(y,labels=mylabels)
plt.show()
```

```python

#Pie chart with labels and colors
y=[10,20,30,40]
mylabels=["Apples","Bananas","Cherries","Dates"]
mycolors=["black","blue","red","hotpink"]
myexplode=[0.2,0,0,0]
plt.pie(y,labels=mylabels, colors=mycolors,explode=myexplode,shadow="true")
plt.show()
```

```python
import numpy as np
x=np.array(["A",'B','C','D'])
y=np.array([3,8,1,10])
mycolors=["black","blue","red","hotpink"]
plt.bar(x,y,color=mycolors)
plt.show()
```

```python
x=np.array(["A",'B','C','D'])
y=np.array([3,8,1,10])
mycolors=["b","g","y","orange"]
plt.barh(x,y,color=mycolors)
plt.show()
```

```python

x=[1,2,3,4,5]
y=[2,4,5,6,7]
colors=["red","blue","pink","black","y"]
sizes=[100,200,300,40,15]
plt.scatter(x,y,c=colors,s=sizes,alpha=0.8,marker="*",edgecolor="g")
plt.title("scatter",fontsize=20)
plt.xlabel("x axis", fontsize=20)
plt.ylabel("y axis",fontsize=10)
plt.show()
```

```python
x=[1,2,3,4]
y=[4,5,6,7]
plt.subplot(2,2,1)
plt.plot(x,y,color="r")
plt.subplot(2,2,2)
plt.pie([2],colors="r")
plt.subplot(2,2,3)
y=[20,40,50,60]
plt.pie(y)
x=np.array(["A",'B','C','D'])
y=np.array([3,8,1,10])
plt.subplot(2,2,4)
plt.bar(x,y)
plt.show()
```
